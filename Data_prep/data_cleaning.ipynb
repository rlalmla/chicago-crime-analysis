## DATA EXPLORATION 


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')   

pd.set_option('display.max_columns', None)
```

```python

file_path = '../Data/raw_data.csv'

df = pd.read_csv(file_path, low_memory=False)

print(f"Dataset loaded: {len(df):,} rows x {len(df.columns)} columns")
print(f"Memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
# Column info
df.info()
```

```python
# First few rows
df.head()
```

```python
# Missing values summary
missing = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum().values,
    'Missing_%': (df.isnull().sum().values / len(df) * 100).round(2)
})

missing = missing[missing['Missing_Count'] > 0].sort_values('Missing_%', ascending=False)
print(missing.to_string(index=False))
```

```python
# Numerical columns summary
df.describe()
```

```python
# Duplicate rows
duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates:,}")
print(f"Percentage: {(duplicates/len(df)*100):.2f}%")
```

```python
# Check rows with missing coordinates
missing_coords = df[df['Latitude'].isnull()]

print(f"Rows missing coordinates: {len(missing_coords):,}")
print(f"\nSample of rows with missing coordinates:")
print(missing_coords[['Case Number', 'Date', 'Primary Type', 'Block', 'Latitude', 'Longitude']].head(10))
```

## DATA CLEANING

```python
# Check if we should keep or remove these rows
print("Crime types with missing coordinates:")
print(missing_coords['Primary Type'].value_counts().head(10))
```

```python
# Save original count
original_count = len(df)

# Remove rows with missing coordinates
df_cleaned = df.dropna(subset=['Latitude', 'Longitude', 'X Coordinate', 'Y Coordinate', 'Location'])

removed = original_count - len(df_cleaned)
print(f"✓ Removed {removed:,} rows with missing coordinates")
print(f"✓ Remaining rows: {len(df_cleaned):,}")
print(f"✓ Retention rate: {(len(df_cleaned)/original_count*100):.2f}%")
```

```python
# Check missing location descriptions
missing_loc_desc = df_cleaned[df_cleaned['Location Description'].isnull()]
print(f"Rows with missing Location Description: {len(missing_loc_desc):,}")

# Fill with 'UNKNOWN'
df_cleaned['Location Description'] = df_cleaned['Location Description'].fillna('UNKNOWN')

print(f"\n✓ Filled missing Location Description with 'UNKNOWN'")
```

```python
# Check remaining missing values
print("Missing Ward:", df_cleaned['Ward'].isnull().sum())
print("Missing Community Area:", df_cleaned['Community Area'].isnull().sum())

# These are very few - we can drop them or fill with mode
# Option: Drop rows with missing Ward/Community Area
df_cleaned = df_cleaned.dropna(subset=['Ward', 'Community Area'])

print(f"\n✓ Removed rows with missing Ward/Community Area")
print(f"✓ Final row count: {len(df_cleaned):,}")
```

```python
# Convert Date column
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], format='%m/%d/%Y %I:%M:%S %p')

print("✓ Date column converted to datetime")
print(f"Date range: {df_cleaned['Date'].min()} to {df_cleaned['Date'].max()}")
```

```python
# Convert Updated On column
df_cleaned['Updated On'] = pd.to_datetime(df_cleaned['Updated On'], format='%m/%d/%Y %I:%M:%S %p')

print("✓ Updated On column converted to datetime")
```

```python
# Extract useful date components
df_cleaned['Month'] = df_cleaned['Date'].dt.month
df_cleaned['Day'] = df_cleaned['Date'].dt.day
df_cleaned['Hour'] = df_cleaned['Date'].dt.hour
df_cleaned['DayOfWeek'] = df_cleaned['Date'].dt.dayofweek  # 0=Monday, 6=Sunday
df_cleaned['DayName'] = df_cleaned['Date'].dt.day_name()

print("✓ Extracted date features: Month, Day, Hour, DayOfWeek, DayName")
df_cleaned[['Date', 'Month', 'Day', 'Hour', 'DayOfWeek', 'DayName']].head()
```

```python
# Check updated data types
print("Updated Data Types:")
print(df_cleaned.dtypes)
```

```python
# Verify no critical missing values remain
missing_check = df_cleaned.isnull().sum()
print("Remaining missing values:")
print(missing_check[missing_check > 0])

if missing_check.sum() == 0:
    print("\n✓ ✓ ✓ NO MISSING VALUES! Dataset is clean.")
else:
    print(f"\n⚠ Still have {missing_check.sum()} missing values to address")
```

```python
print("="*60)
print("STEP 2 SUMMARY")
print("="*60)
print(f"Original dataset: {original_count:,} rows")
print(f"Cleaned dataset: {len(df_cleaned):,} rows")
print(f"Rows removed: {original_count - len(df_cleaned):,}")
print(f"Data retention: {(len(df_cleaned)/original_count*100):.2f}%")
print(f"\nNew features added: Month, Day, Hour, DayOfWeek, DayName")
print(f"Total columns: {len(df_cleaned.columns)}")
```

```python
# Check current state of text columns
text_columns = ['Case Number', 'Block', 'IUCR', 'Primary Type', 'Description', 'Location Description', 'FBI Code']

print("Sample values before standardization:")
for col in text_columns[:3]:
    print(f"\n{col}:")
    print(df_cleaned[col].head(3).tolist())
```

```python
# Standardize: remove leading/trailing spaces and multiple spaces
for col in text_columns:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)

print("✓ Standardized text columns (removed extra spaces)")
```

```python
# Primary Type - Crime categories
print("Unique Primary Types:", df_cleaned['Primary Type'].nunique())
print("\nTop 15 Crime Types:")
print(df_cleaned['Primary Type'].value_counts().head(15))
```

```python
# Location Description
print("Unique Location Descriptions:", df_cleaned['Location Description'].nunique())
print("\nTop 15 Locations:")
print(df_cleaned['Location Description'].value_counts().head(15))
```

```python
# FBI Code
print("Unique FBI Codes:", df_cleaned['FBI Code'].nunique())
print("\nFBI Code Distribution:")
print(df_cleaned['FBI Code'].value_counts())
```

```python
# Check Arrest and Domestic columns
print("Arrest Distribution:")
print(df_cleaned['Arrest'].value_counts())
print(f"Arrest Rate: {(df_cleaned['Arrest'].sum() / len(df_cleaned) * 100):.2f}%")

print("\nDomestic Distribution:")
print(df_cleaned['Domestic'].value_counts())
print(f"Domestic Crime Rate: {(df_cleaned['Domestic'].sum() / len(df_cleaned) * 100):.2f}%")
```

```python
# Check Year column
print("Year Distribution:")
print(df_cleaned['Year'].value_counts().sort_index())

# Validate Year matches extracted year from Date
df_cleaned['Year_from_Date'] = df_cleaned['Date'].dt.year
year_mismatch = df_cleaned[df_cleaned['Year'] != df_cleaned['Year_from_Date']]

print(f"\nYear mismatches: {len(year_mismatch)}")
if len(year_mismatch) > 0:
    print("Sample mismatches:")
    print(year_mismatch[['Date', 'Year', 'Year_from_Date']].head())
else:
    print("✓ All Year values match Date column")

# Drop temporary column
df_cleaned = df_cleaned.drop('Year_from_Date', axis=1)
```

```python

# Check District, Ward, Community Area ranges
print("District range:", df_cleaned['District'].min(), "to", df_cleaned['District'].max())
print("Number of unique Districts:", df_cleaned['District'].nunique())

print("\nWard range:", df_cleaned['Ward'].min(), "to", df_cleaned['Ward'].max())
print("Number of unique Wards:", df_cleaned['Ward'].nunique())

print("\nCommunity Area range:", df_cleaned['Community Area'].min(), "to", df_cleaned['Community Area'].max())
print("Number of unique Community Areas:", df_cleaned['Community Area'].nunique())
```

```python
# Chicago approximate boundaries
# Latitude: ~41.64 to 42.02
# Longitude: ~-87.94 to -87.52

lat_min, lat_max = 41.6, 42.1
lon_min, lon_max = -88.0, -87.5

print("Checking coordinates are within Chicago bounds...")
print(f"Latitude range: {df_cleaned['Latitude'].min():.6f} to {df_cleaned['Latitude'].max():.6f}")
print(f"Longitude range: {df_cleaned['Longitude'].min():.6f} to {df_cleaned['Longitude'].max():.6f}")

# Find outliers
outlier_coords = df_cleaned[
    (df_cleaned['Latitude'] < lat_min) | (df_cleaned['Latitude'] > lat_max) |
    (df_cleaned['Longitude'] < lon_min) | (df_cleaned['Longitude'] > lon_max)
]

print(f"\nCoordinate outliers: {len(outlier_coords)}")
if len(outlier_coords) > 0:
    print("\nSample outliers:")
    print(outlier_coords[['Block', 'Latitude', 'Longitude', 'Primary Type']].head())
else:
    print("✓ All coordinates within Chicago bounds")
```

```python
# Check if Updated On is after Date (should be)
future_updates = df_cleaned[df_cleaned['Updated On'] < df_cleaned['Date']]

print(f"Records where 'Updated On' is before 'Date': {len(future_updates)}")
if len(future_updates) > 0:
    print("\nSample:")
    print(future_updates[['Case Number', 'Date', 'Updated On']].head())
else:
    print("✓ All update dates are logical")
```

```python
# Check for invalid hour values (should be 0-23)
invalid_hours = df_cleaned[(df_cleaned['Hour'] < 0) | (df_cleaned['Hour'] > 23)]

print(f"Invalid hour values: {len(invalid_hours)}")
if len(invalid_hours) == 0:
    print("✓ All hours are valid (0-23)")

# Distribution by hour
print("\nCrime distribution by hour:")
print(df_cleaned['Hour'].value_counts().sort_index())
```

```python
# Convert Ward and Community Area to integers (they're currently float)
df_cleaned['Ward'] = df_cleaned['Ward'].astype('int32')
df_cleaned['Community Area'] = df_cleaned['Community Area'].astype('int32')

# Convert smaller integer columns to save memory
df_cleaned['Beat'] = df_cleaned['Beat'].astype('int16')
df_cleaned['District'] = df_cleaned['District'].astype('int8')
df_cleaned['Year'] = df_cleaned['Year'].astype('int16')
df_cleaned['Month'] = df_cleaned['Month'].astype('int8')
df_cleaned['Day'] = df_cleaned['Day'].astype('int8')
df_cleaned['Hour'] = df_cleaned['Hour'].astype('int8')
df_cleaned['DayOfWeek'] = df_cleaned['DayOfWeek'].astype('int8')

print("✓ Optimized data types for memory efficiency")
```

```python
# Memory comparison
memory_mb = df_cleaned.memory_usage(deep=True).sum() / 1024**2
print(f"Current memory usage: {memory_mb:.2f} MB")
print(f"\nMemory by column:")
mem_by_col = df_cleaned.memory_usage(deep=True).sort_values(ascending=False)
print(mem_by_col.head(10))
```

```python
print("="*60)
print("CLEANED DATASET SUMMARY")
print("="*60)
print(f"Total Rows: {len(df_cleaned):,}")
print(f"Total Columns: {len(df_cleaned.columns)}")
print(f"Memory Usage: {df_cleaned.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"\nDate Range: {df_cleaned['Date'].min()} to {df_cleaned['Date'].max()}")
print(f"\nMissing Values: {df_cleaned.isnull().sum().sum()}")
print(f"\nData Types:")
print(df_cleaned.dtypes.value_counts())
```
